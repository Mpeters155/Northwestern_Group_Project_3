```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
# %%
drop_cols = ['Open', 'High', 'Low','Volume', 'Dividends', 'Stock Splits', 'Capital Gains']
```

```python
def get_history(fund_ticker):
    tckr_name = yf.Ticker(fund_ticker)
    ticker_df = tckr_name.history(period = '10y')
    ticker_df = ticker_df.drop(columns= drop_cols)
    ticker_df = ticker_df.dropna()
    ticker_df.index = ticker_df.index.date
    ticker_df.index.rename('Date', inplace = True)
    ticker_df.reset_index(inplace=True)
    ticker_df.rename(columns = {"Close": "Close Price"}, inplace = True)
    return ticker_df
```

```python
ticker = input(str('What Mutual Fund Would you like?'))
```

```python
ticker_df = get_history(ticker)
```

```python
prophet_df = ticker_df[['Date','Close Price']]
```

```python
prophet_df.head()
```

```python
prophet_df = prophet_df.rename(columns = {'Date':'ds','Close Price':'y'})
```

```python
prophet_df.head()
```

```python
from prophet import Prophet
# The Prophet class (model)
fbp = Prophet(daily_seasonality = True)
 # Fit the model
fbp.fit(prophet_df)
```

```python
# We need to specify the number of days in future
fut = fbp.make_future_dataframe(periods=730)
forecast = fbp.predict(fut)
```

```python

from prophet.plot import plot_plotly, plot_components_plotly
# A better plot than the simple matplotlib
plot_plotly(fbp, forecast)
```
